```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
xL=np.array([3,6,9,15])
yL=np.array([1,7,10,12])
X,Y= np.meshgrid(xL,yL)
```

```python
Z=np.random.randint(1,3,(4,4))
```

```python
plt.contour(X,Y,Z)
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contourf(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```
