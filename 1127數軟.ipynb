```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt

import pandas as pd
```

## PCA

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.decomposition import PCA
```

## step 1.開PCA機

```python
pca=PCA(n_components=2)
```

### step 2.訓練

```python
pca.fit(x)
```

## 3.預測

```python
x[87]
```

```python
X=pca.transform(x)
```

```python
x[87]
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

## 準備訓練/測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

## step 1:開SVC函數學習機

```python
from sklearn.svm import SVC
```

```python
clf=SVC(gamma='auto')
```

## step2: 訓練 

```python
clf.fit(x_train,y_train)
```

## step3:預測

```python
y_predict=clf.predict(x_test)
```

```python
y_predict
```

```python
x0=np.arange(-4,4.2,0.02)
y0=np.arange(-1.5,1.7,0.02)
xm,ym =np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)
plt.contourf(xm,ym,Z, alpha=0.3,cmap='Paired')
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,cmap='Paired')
```

## K-Means自動分類

```python
x=np.random.rand(100,2)
```

```python
plt.scatter(x[:,0],x[:,1])
```

## Step 1.打開K-Means學習機

```python
from sklearn.cluster import KMeans
```

```python
clf=KMeans(n_clusters=3)
```

## step 2.學習

```python
clf.fit(x)
```

## 3.預測

```python
clf.predict(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```

```python
x0=y0=np.arange(-0.2,1.2,0.02)

xm,ym =np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)
plt.contourf(xm,ym,Z, alpha=0.3,cmap='Paired')
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,cmap='Paired')
```

## Means Shifts完全自動分類

### step 1. 打開函數學習機

```python
from sklearn.cluster import MeanShift
```

```python
clf=MeanShift(bandwidth=0.9)
```

## step 2. 訓練

```python
clf.fit(x)
```

## step 3. 預覽

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

```python
x0=np.arange(-3.8,5.2,0.02)
y0=np.arange(-1.8,7,0.02)
xm,ym =np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)
plt.contourf(xm,ym,Z, alpha=0.3,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```
